```python
import pandas as pd

# Use the exact path found by your search loop
path = '/kaggle/input/datasets/shubhra2001/hotel-booking-dataset/hotel_bookings.csv'
df = pd.read_csv(path)

# Preview the first 5 rows
df.head()
```

```python
df.info()
df.describe()
```

# children → small missing

```python
# fill missing values (correct way)
df['children'] = df['children'].fillna(0)
df['agent'] = df['agent'].fillna(0)
df['country'] = df['country'].fillna('Unknown')
```

```python
# fill missing values
df['children'] = df['children'].fillna(0)
df['agent'] = df['agent'].fillna(0)
df['country'] = df['country'].fillna('Unknown')

# drop company safely
df = df.drop('company', axis=1, errors='ignore')

# remove invalid adr
df = df[df['adr'] >= 0]

# remove duplicates
df = df.drop_duplicates()
```

# Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='hotel', data=df)
plt.title("Hotel Type Distribution")
plt.show()

sns.countplot(x='is_canceled', data=df)
plt.title("Cancellation Count")
plt.show()
```

# cancellation rate by hotel

```python
df.groupby('hotel')['is_canceled'].mean()
```
